Heapify

Recall that to build a binary search tree with n elements, the time complexity is O(n log n), because each insertion takes O(log n) time and there are n elements to insert.

If we build our heap of size n by pushing each element one-by-one, this would also run in O(n log n) time. But there is actually a more efficient algorithm known as heapify, which allows us to perform this operation in O(n) time.

To build a heap we must satisfy the heap structure and order properties.

The structure property is trivial to satisfy because we can initialize an empty array and append elements to it. The order property is a bit more complex.

The order property states that the parent node must be less than or equal to its children. We can accomplish this by percolating each element down the tree until it reaches the correct position. Since the leaf nodes cannot be shifted any lower, we can skip the last level of the tree.

1. We calculate the index of the first non-leaf node, which is n/2 , where n is the number of elements in the heap.
2. Starting from the first non-leaf node, we will percolate down, the exact same way we did in the pop() function.
3. After each iteration, we are going to decrement the index by 1 so we can perform percolate down on the next non-leaf node.
4. We will continue this process until we reach the root node.

```python
def heapify(self, arr):
    # 0-th position is moved to the end
    arr.append(arr[0])

    self.heap = arr
    cur = (len(self.heap) - 1) // 2
    while cur > 0:
        # Percolate down
        i = cur
        while 2 * i < len(self.heap):
            if (2 * i + 1 < len(self.heap) and 
            self.heap[2 * i + 1] < self.heap[2 * i] and 
            self.heap[i] > self.heap[2 * i + 1]):
                # Swap right child
                tmp = self.heap[i]
                self.heap[i] = self.heap[2 * i + 1]
                self.heap[2 * i + 1] = tmp
                i = 2 * i + 1
            elif self.heap[i] > self.heap[2 * i]:
                # Swap left child
                tmp = self.heap[i]
                self.heap[i] = self.heap[2 * i]
                self.heap[2 * i] = tmp
                i = 2 * i
            else:
                break
        cur -= 1
```

Time Complexity
Given that there are n nodes in a binary tree, there are roughly n / 2 leaf nodes and n / 2 non-leaf nodes.

We only percolate down the non-leaf nodes

In the example, the nodes on the 2nd level need to percolate down one level, and the nodes on the 1st level are percolating down two levels, with the root node having to percolate down all the levels. So while the number of nodes is halving each time, the number of levels needed to be percolated increases. There is a very neat mathematical summation that is formed when we add all the work together, which simplifies to O(n).

We will not be discussing the proof behind this since it is highly mathematical and it's unlikely that you will be asked to prove it in an interview.

Time Complexity:

```python
# 1046. Last Stone Weight
# Input: stones = [2,7,4,1,8,1]
# Output: 1

def lastStoneWeight(self, stones: List[int]) -> int:
    # normally its minHeap, so use -1 to implement maxHeap 
    stones = [-i for i in stones]

    heapq.heapify(stones)

    while len(stones) > 1:
        s1 = -1 * heapq.heappop(stones)
        s2 = -1 * heapq.heappop(stones)

        if s1 - s2 > 0: 
            heapq.heappush(stones, -1*(s1 - s2))
    
    return -1*stones[0] if len(stones) == 1 else 0

# Time O(n log n)

# stones = [-i for i in stones]: O(n) - iterate through all stones
# heapq.heapify(stones): O(n) - heapify is linear time
# While loop: Runs at most (n-1) times (each iteration removes at least one stone)
    # 2 × heappop: O(log n) each → O(2 log n)
    # 1 × heappush (sometimes): O(log n)
    # Per iteration: O(log n)
    # Total for loop: O(n log n)
# Overall: O(n) + O(n) + O(n log n) = O(n log n)
```

```python
# 973. K Closest Points to Origin
# Input: points = [[3,3],[5,-1],[-2,4]], k = 2
# Output: [[3,3],[-2,4]]

def kClosest(self, points: List[List[int]], k: int) -> List[List[int]]:
    # find distance to the origin for each point
    minHeap = []
    
    for x, y in points: # O(n)
        distance = pow(x, 2) + pow(y, 2)
        minHeap.append([distance, x, y])

    heapq.heapify(minHeap) # O(n)

    res = []
    while k > 0: 
        distance, x, y = heapq.heappop(minHeap) # O(log n)
        res.append([x,y])
        k -= 1
    
    return res

# O(N+Klogn)
```

```python
# 215. Kth Largest Element in an Array
# Input: nums = [3,2,1,5,6,4], k = 2
# Output: 5

def findKthLargest(self, nums: List[int], k: int) -> int:
    nums = [-n for n in nums]
    heapq.heapify(nums)
    
    res = []
    while k > 0:
        res.append(-1*heapq.heappop(nums))
        k -= 1
    return res[-1]

# Time O(N+Klogn)
```
